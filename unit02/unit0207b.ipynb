```python
from bs4 import BeautifulSoup
import requests, json
import time

data = list()
def fetch(url):  #讀取單一頁面的函式
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")
    selector = "#pageptlist > div > div > div > div.d-txt > div.mtitle > a"
    target = soup.select(selector)
    for item in target:
        element = dict()
        element["title"] = item.text.strip()
        element["link"] = item["href"]
        data.append(element)

url_pattern = "https://www.nkust.edu.tw/p/403-1000-1363-{}.php?Lang=zh-tw"
for page in range(1, 64):
    url = url_pattern.format(page)
    print("Fetching page: {}".format(page))
    fetch(url)
    time.sleep(3)
with open("nkust_news_abs.json", "w", encoding="utf-8") as json_file:
    json.dump(data, json_file, indent=4)
```
